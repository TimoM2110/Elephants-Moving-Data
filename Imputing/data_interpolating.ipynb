```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\timom\anaconda3\envs\madeleine_project\ThermochronTracking Elephants Kruger 2007.csv")
df.tail()
```

```python
tags = [
    'AM105',
    'AM107',
    'AM108',
    'AM110',
    'AM239',
    'AM253',
    'AM254',
    'AM255',
    'AM306',
    'AM307',
    'AM308',
    'AM91',
    'AM93',
    'AM99',
]
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'],errors='ignore')
```

```python
df = df.set_index('timestamp')
```

```python
def imputing(element, freq):
    sf = df.loc[df['tag-local-identifier'] == tags[element]]
    resample_index = pd.date_range(start=sf.index[0], end=sf.index[-1], freq=freq)
    dummy_frame = pd.DataFrame(np.NaN, index=resample_index, columns=sf.columns)
    ds = sf.combine_first(dummy_frame).interpolate('time').resample(freq).asfreq()
    ds['sensor-type']='gps'
    ds['individual-taxon-canonical-name']='Loxodonta africana'
    ds['tag-local-identifier']=tags[element]
    ds['individual-local-identifier']=tags[element]
    ds['study-name']='ThermochronTracking Elephants Kruger 2007'
    ds['visible']='True'
    ds.to_csv(tags[element] +'.csv')
```

```python
for element in range(len(tags)):
    imputing(element, '30min')
```
